```python
##BGR to RGB conversion
import cv2;
import matplotlib.pyplot as plt;
image_path=cv2.imread('./Image/red.jpg');
blue,green,red=cv2.split(image_path);
rgb=cv2.cvtColor(image_path,cv2.COLOR_BGR2RGB);
plt.figure(figsize=(6,10));

plt.subplot(2,2,1);
plt.imshow(rgb);
plt.title('RGB')

plt.subplot(2,2,2);
plt.imshow(blue,cmap='gray');
plt.title('BLUE');

plt.subplot(2,2,3);
plt.imshow(green,cmap='gray');
plt.title('GREEN');

plt.subplot(2,2,4);
plt.imshow(red,cmap='gray');
plt.title('RED');






```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load image
image_path = './Image/red.jpg'
image = cv2.imread(image_path)

# Convert BGR to GRAYSCALE
BGR_TO_GRAY_COLOR = cv2.COLOR_BGR2GRAY
gray_image = cv2.cvtColor(image, BGR_TO_GRAY_COLOR)

# Convert BGR to HSV
BGR_TO_HSV_COLOR = cv2.COLOR_BGR2HSV
hsv_image = cv2.cvtColor(image, BGR_TO_HSV_COLOR)

# Convert BGR to RGB for displaying the original image with plt.imshow
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the images
plt.figure(figsize=(15, 5))

# Plot original image
plt.subplot(1, 3, 1)
plt.imshow(rgb_image)  # Display RGB image
plt.axis('off')
plt.title('Original Image')

# Plot grayscale image
plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')  # Correct cmap for grayscale
plt.axis('off')
plt.title('Grayscale Image')

# Plot HSV image
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB))  # Convert HSV to RGB for proper display
plt.axis('off')
plt.title('HSV Image')

plt.show()












```
